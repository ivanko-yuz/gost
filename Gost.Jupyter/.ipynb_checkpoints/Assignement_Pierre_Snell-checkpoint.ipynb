### 1.5 Represent  the  scatterplots  of  the  three  most  correlated  pairs  of  variables

We make a list of 3 maximums. 
We apply the classical maximum algorithm, when one value is bigger than the minimum of the temporary maximum list, we replace this minimum by the new value.  

```python
maximum = [0,0,0]
ind_max = [[0,0],[0,0],[0,0]]
temp = 0
for i in correlations.columns:
    for j in correlations.index:
        if i < j :                 # We look only bellow the diagonal of the correlation matrix (which is symetric)
            if abs(correlations[i][j])>=min(maximum):
                ind_max[maximum.index(min(maximum))] = [i,j]
                maximum[maximum.index(min(maximum))]=abs(correlations[i][j])
print(maximum)
print(ind_max)
```

```python
for i in range(len(ind_max)):
    x = dataset[ind_max[i][0]]
    y = dataset[ind_max[i][1]]
    
    plt.figure();
    
    plt.scatter(x,y)

    reg_line = np.polyfit(x, y, 1)
    fct_reg_line = np.poly1d(reg_line)
    
    plt.plot(x,fct_reg_line(x),"b--")

    plt.title("Index "+str(ind_max[i])+"\n Correlation : "+str(maximum[i]))
    
    plt.show()
```

**We can see that the most correlated is the third one, more close to the regression line.**

**The second is spreaded around the regression line, it's why the coefficient is only of 0.6.**

# *Question n°2* : Compute sample  mean  m and sample  standard  deviation  s for  all  variables.

$$(1) ~~\sigma = \sqrt{\frac{\sum_{i=1}^{n}{(x_{i}-m)^{2}}}{n}}$$
$$ (2) ~~ m = \frac{\sum_{i=1}^{n}{x_{i}}}{n}$$ 

We compute the mean, the classical way and then we compare to a reliable method (panda.mean())

```python
m = pd.DataFrame(dataset.sum()/dataset.shape[0]).T
print("Own average minus pandas one: \n")
print(m-dataset.mean())
print("\n The median by columns")
m
```

<font color=blue> **We apply the formula (1) ** </font>

```python
var = pd.Series(index=dataset.columns)

for i in dataset.columns:
    xc = dataset[i].apply(lambda x : x-m[i])        # We center the element of each columns 
    xc2 = xc.apply(lambda x : pow(x,2))             # Then we square them 
    sigma2 = xc2.sum()                              # And we sum them 
    var[i] = (sigma2/xc2.shape[0]).apply(np.sqrt)   # Then we divide by n and we take the square root 
  
```

**Own Variance**

**Error between own variance and panda.DataFrame.std() **

```python
pd.DataFrame(var-dataset.std(ddof=1)).T
```

**Error in %  ** 

```python
err = var-dataset.std(ddof=1)
pd.DataFrame((err/dataset.std(ddof=1))*100).T
```

# *Question n°3* : Write  a python function  that  computes confidence interval for  the  population  mean under  the  assumption  that  all  are  normally  distributed

$$(3) ~~ [~\bar{x} - \alpha\frac{\sigma(X)}{\sqrt{n}}~;~\bar{x} + \alpha\frac{\sigma(X)}{\sqrt{n}}~] $$

We compute the intervals by two ways :

(1) The empirical one, taking many samples of our dataset and computing their means. These means follow a normal distribution if we take enough of them. 

(2) By the "theoretical" method by applying the formula (3)  with $\alpha$ the quantile of the normal distribution of the order we want for the computation  
 

```python
# Confidence interval, computed with the formula (3)
def conf_inter_theo(dataset,alpha):
    N = dataset.shape[0]
    sigma = dataset.std()
    coef = sp.norm.ppf(1-alpha/2)   # the quantile of order alpha we want computed by scypy.stat
    moy = dataset.mean()
    return pd.DataFrame([moy-coef*(sigma/np.sqrt(N)),moy+coef*(sigma/np.sqrt(N))],index=[alpha/2,1-alpha/2])
        


# Confidence interval computed empiricaly
# we run a epoch number of iteration by taking a sample of size batch among our dataset
def conf_inter_emp(dataset, alpha,p=0):
    batch = dataset.shape[0]-dataset.shape[0]//10    # Size of a column minus 10% of the size of the column
    epoch = 1000
    lower = alpha / 2
    upper = 1-alpha / 2

    avg = pd.DataFrame(dataset.sample(batch,replace = True).mean()).T   # Initialisation to be able to append. 

    for _ in range(epoch):                                      # for _ means for "nothing" just do it epoch times
        data_to_mean = (dataset.sample(batch,replace = True).mean()).T
        avg = avg.append(data_to_mean,ignore_index=True)                # We place all the sample mean in an array
                                                                # This means will follow a normal distribution
    if (p==1):
        avg.hist()
    return avg.quantile([lower, upper])
```

## With other tools 

```python
def conf_inter(dataset,alpha):
    conf_int = pd.DataFrame(index = [alpha/2 , 1-alpha/2])
    for i in dataset.columns: 
        conf_int[i] = sp.norm.interval(alpha/2, loc=dataset[i].mean(), scale=dataset[i].std()/dataset[i].shape[0])
    return conf_int
```

```python
conf_inter(dataset,0.05)
```

```python
conf_inter(dataset,0.01)
```

**The intervals founds are really narrow, but still included in those found with our own method. This is surely due by the way sp.norm.interval() compute the intervals (perfect distribution with infinite limits.) **

# *Question n°7*  : Perform Principal  Component  Analysison  the  individuals  with  scaling.  Produce  the biplot  and  the  scatterplot  of  units  on  the  factorial  plane.  Comment  on  the  %  of variance  explained,  on  collinearity  between  variables,  on  the  main  variables  that influence  the  interpretation  of  the  factorial  plane

```python
dataset_std = pd.DataFrame(StandardScaler().fit_transform(dataset))
cov_std = dataset_std.corr()
eig_vals, eig_vect = np.linalg.eig(cov_std)
eig_pairs = [(np.abs(eig_vals[i]), eig_vect[:,i]) for i in range(len(eig_vals))]
eig_pairs.sort()
eig_pairs.reverse()
sum_ev = sum(eig_vals)
pve = [(i / sum_ev)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_pve = np.cumsum(pve)

fig = plt.figure(figsize=[10,5])
plt.scatter(['PC%s' %i for i in range(1,15)],pve,s=80)
plt.scatter(['PC%s' %i for i in range(1,15)],cum_var_pve,marker='+')
plt.legend(['Variance','Cumulative variance'])
plt.show()
```

**We need to take components with the highest value to keep the information on the projection space. Here we're sure that we need the first and the second. For the rest we run the computation bellow. **

```python
pca = PCA().fit(dataset)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative variance');
```

**We can see that 2 components is the good ration between the size of the projection space and the loss of information.** 

**We could have taken the eig_vect from last cell, order them and take the two most important. Then put them in a matrix and compute the dot product with our dataset to get the projection but the sklearn method is faster, more reliable and give directly the points to plot. **

```python
dataset_pca = PCA(n_components=2).fit_transform(dataset_std)
```

# *Question n°8* : Cluster  the  individuals  using K-means for k = 2,  .  .  .  ,  20.  Choose  the  ideal  number of  clusters  according  to  a  method  of  your  choice  and  provide  motivations  for  your choice.  Exhibit  and  comment  the  silhouette  plot  of  the  chosen  clustering.  Provide the  plot  of  the  clustered  points  on  the  factorial  plane

```python
k = range(2,21)
silhouette = [0.0]*21
for n_clusters in k:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(dataset_pca)
    silhouette_avg = silhouette_score(dataset_pca, cluster_labels)
    silhouette[n_clusters] = silhouette_avg

best_nb_clust = silhouette.index(max(silhouette)) # We compute the score for each cluster and take the closest to 1
print("The best number of cluster is : "+str(best_nb_clust))
```

```python
kmeans_label = KMeans(n_clusters=best_nb_clust, random_state=10).fit_predict(dataset_pca)
plt.scatter(dataset_pca[:, 0], dataset_pca[:, 1], c=kmeans_label,s=50,cmap='viridis');
```

```python
X = dataset_pca
range_n_clusters = range(2,21)

for n_clusters in range_n_clusters:
    fig, (ax1, ax2) = plt.subplots(1, 2)              # Subplot : 1 row ,2 columns
    fig.set_size_inches(18, 7)                        # The 1st subplot is the silhouette plot
    
    ax1.set_xlim([-0.2, 1])                             # Limit of the figure for the silhouette -1, 1 
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10]) # The (n_clusters+1)*10 is for blank space between silhouette

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)  # Silhouette score between -1 (worse) and 1 (better) 
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = plt.cm.inferno(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.2, -0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = plt.cm.inferno(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c= cluster_labels , edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

    plt.show()
```

**We can see that, indeed, the best silhouette is with n_cluster = 2, after, clustering more lead to more peaks and even negatives values.** 

**Perhaps with an other normalisation and/or other projections we can find more properties of this column of the dataset. Here it's still compact without interesting informations. **
